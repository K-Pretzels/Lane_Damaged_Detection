```python
import cv2
import numpy as np
from ultralytics import YOLO
from ultralytics.utils.plotting import Annotator, colors
from collections import defaultdict

track_history = defaultdict(lambda: [])

model = YOLO("./segment_best.pt")  # segmentation model
cap = cv2.VideoCapture("Z:/동영상/2022-08-04T00_01_50.485Z_0.mp4")
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

while True:
    ret, im0 = cap.read()
    if not ret:
        print("Video frame is empty or video processing has been successfully completed.")
        break

    im0 = cv2.resize(im0, (int(w / 2), int(h / 2)))

    annotator = Annotator(im0, line_width=2)

    results = model.track(im0, persist=True)

    try:
        if results[0].boxes.id is not None and results[0].masks is not None:
            masks = results[0].masks.xy
            track_ids = results[0].boxes.id.int().cpu().tolist()

            for mask, track_id in zip(masks, track_ids):
                # 다각형 마스크 생성 및 이미지 추출
                mask_np = np.array(mask, dtype=np.int32)  # xy 좌표를 numpy 배열로 변환
                mask_img = np.zeros(im0.shape[:2], dtype=np.uint8)
                cv2.fillPoly(mask_img, [mask_np], 255)
                result = cv2.bitwise_and(im0, im0, mask=mask_img)

                non_black_pixels = np.where((result[:, :, 0] != 0) | (result[:, :, 1] != 0) | (result[:, :, 2] != 0))
                # 픽셀 값이 150 이상인 것들의 비율 계산
                over_120_pixels = np.where((result[:, :, 0] >= 120) & (result[:, :, 1] >= 120) & (result[:, :, 2] >= 120))
                percentage_over_120 = len(over_120_pixels[0]) / len(non_black_pixels[0]) * 100
                
                annotator.seg_bbox(mask=mask,
                                   mask_color=colors(track_id, True),
                                   track_label=str(track_id)+f' / {percentage_over_120:.2f}%')
                
                # 결과 이미지 저장
                #cv2.imwrite(f'track_{track_id}.png', result)

    except Exception as e:
        print(f"Error: {e}")
        pass

    cv2.imshow("instance-segmentation-object-tracking", im0)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
from ultralytics.utils.plotting import Annotator, colors
from collections import defaultdict

track_history = defaultdict(lambda: [])

model = YOLO("./segment_best.pt")  # segmentation model
cap = cv2.VideoCapture("Z:/동영상/2022-08-02T06_59_40.135Z_1.mp4")
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

while True:
    ret, im0 = cap.read()
    if not ret:
        print("Video frame is empty or video processing has been successfully completed.")
        break

    im0 = cv2.resize(im0, (int(w / 2), int(h / 2)))

    annotator = Annotator(im0, line_width=2)

    results = model.track(im0, persist=True)

    try:
        if results[0].boxes.id is not None and results[0].masks is not None:
            masks = results[0].masks.xy
            track_ids = results[0].boxes.id.int().cpu().tolist()

            for mask, track_id in zip(masks, track_ids):
                mask_np = np.array(mask, dtype=np.int32)  # xy 좌표를 numpy 배열로 변환
                mask_img = np.zeros(im0.shape[:2], dtype=np.uint8)
                cv2.fillPoly(mask_img, [mask_np], 255)

                # 윤곽선 찾기
                contours, _ = cv2.findContours(mask_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
                # 윤곽선을 이용하여 관심 영역만 포함하는 새로운 이미지 생성
                mask_img_color = cv2.cvtColor(mask_img, cv2.COLOR_GRAY2BGR)  # 마스크를 컬러 이미지로 변환
                cv2.drawContours(mask_img_color, contours, -1, (0, 255, 0), 3)
                result = cv2.bitwise_and(im0, mask_img_color)

                # 결과 이미지 저장
                cv2.imwrite(f'track_{track_id}_segmented.png', result)


    except Exception as e:
        print(f"Error: {e}")
        pass

    cv2.imshow("instance-segmentation-object-tracking", im0)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
